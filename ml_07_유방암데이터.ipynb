```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 데이터 불러오기

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.target_names
```

```python
np.bincount(cancer.target)
```

```python
cancer.data.shape
```

```python
cancer.feature_names
```

```python
for i,fn in enumerate(cancer.feature_names):
    print(i,'\t',fn)
```

### 산점도

```python
plt.figure(figsize=[12,16])

for i in range(30):
    plt.subplot(5,6,i+1)
    plt.scatter(cancer.data[:,0],cancer.data[:,i], c=cancer.target,
                cmap='autumn', alpha=0.1)
    plt.title(cancer.feature_names[i])
    plt.axis('off')
```

### 히스토그램

```python
malignant = cancer.data[cancer.target==0]
benign = cancer.data[cancer.target==1]
malignant.shape, benign.shape
```

```python
fig=plt.figure(figsize=[14,14])
fig.suptitle('Breast Cancer - feature analysis', fontsize=20)

for col in range(30):
    plt.subplot(8,4,col+1)
    _,bins=np.histogram(cancer.data[:,col],bins=50)
    plt.hist(malignant[:,col], bins=bins, alpha=0.5, label='malignant', color='red')
    plt.hist(benign[:,col], bins=bins, alpha=0.5, label='benign', color='green')
    
    plt.title(cancer.feature_names[col]+('(%d)' % col))
    plt.xticks([])
    plt.yticks([])
    if col==0: plt.legend()
```
